```python
from sklearn.preprocessing import LabelEncoder

encoder=LabelEncoder()

df['sentiment']=encoder.fit_transform(df['sentiment'])
df['sentiment']
```

```python
#.values is used to convert Pandas Series into Numpy Array(which works best with keras)
X=df['review'].values
y=df['sentiment'].values
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
VOCAB_SIZE=10000
MAX_LEN=200
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenizer=Tokenizer(num_words=VOCAB_SIZE)
tokenizer.fit_on_texts(X_train)
```

```python
X_train_sequence=tokenizer.texts_to_sequences(X_train)
X_test_sequence=tokenizer.texts_to_sequences(X_test)
```

```python
X_train_pad=pad_sequences(X_train_sequence,maxlen=MAX_LEN)
X_test_pad=pad_sequences(X_test_sequence,maxlen=MAX_LEN)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten,Embedding

model=Sequential()

model.add(Embedding(input_dim=VOCAB_SIZE,output_dim=64,input_length=MAX_LEN))

model.add(Flatten())

model.add(Dense(64,activation='relu'))

model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

model.summary()
```

```python
history=model.fit(X_train_pad,y_train,epochs=10,batch_size=32,validation_split=0.1,verbose=1)
```

```python
loss, accuracy=model.evaluate(X_test_pad,y_test)
print("Loss:",loss)
print("Accuracy:",accuracy)
```

```python
y_pred_prob=model.predict(X_test_pad)
y_pred=(y_pred_prob>0.5).astype(int)
```

```python
from sklearn.metrics import confusion_matrix,classification_report

cm=confusion_matrix(y_test,y_pred)
print("Confusion Matrix:",cm)

print("\n Classification Report:",classification_report(y_test,y_pred))
```
